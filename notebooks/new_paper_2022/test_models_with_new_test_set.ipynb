```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score, \
    balanced_accuracy_score, precision_score

import numpy as np

y_predict = []
for prediction in predictions:
    if isinstance(prediction, np.ndarray):
        if prediction[1] > 0.5:
            y_predict.append(1)
        else:
            y_predict.append(0)
    else:
        if prediction > 0.5:
            y_predict.append(1)
        else:
            y_predict.append(0)

print(confusion_matrix(y_true, y_predict))
print(classification_report(y_true, y_predict))
print(f1_score(y_true, y_predict))
print(balanced_accuracy_score(y_true, y_predict))
print(precision_score(y_true, y_predict))
```

```python
import pandas as pd
df = pd.read_csv("/home/joao/Desktop/DeepSweet/notebooks/new_paper_2022/bittersweetforest-test.csv")
molecules = df["Canonical SMILES"].tolist()
print(len(molecules))

predictions, dataset = ensemble.predict(molecules)
```

```python
y_true = df.iloc[dataset.ids, :].Sweet.replace({True: 1,False: 0}).values
molecules = dataset.mols
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, f1_score, \
    balanced_accuracy_score, roc_auc_score,  precision_score

y_predict = []
fp = []
fn = []

for i, prediction in enumerate(predictions):
    if isinstance(prediction, list) or isinstance(prediction, np.ndarray):
        if prediction[1] > 0.54:
            if y_true[i] != 1:
                fp.append((molecules[i], prediction[1]))
            y_predict.append(1)
        else:
            if y_true[i] != 0:
                fn.append((molecules[i], prediction[1]))
            y_predict.append(0)
    else:
        if prediction > 0.54:
            if y_true[i] != 1:
                fp.append((molecules[i], prediction))
            y_predict.append(1)
        else:
            if y_true[i] != 0:
                fn.append((molecules[i], prediction))
            y_predict.append(0)

print(confusion_matrix(y_true, y_predict))
print(classification_report(y_true, y_predict))
print(roc_auc_score(y_true, y_predict))
print(f1_score(y_true, y_predict))
print(balanced_accuracy_score(y_true, y_predict))
print(precision_score(y_true, y_predict))
```
